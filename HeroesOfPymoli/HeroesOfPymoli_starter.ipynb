```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Drop duplicates on the 'Series Name' series, perform a count and assign it to a new Data Frame 
players_count = purchase_data['SN'].drop_duplicates().count()
players_count_df = pd.DataFrame({'Total Players':[players_count]})
players_count_df           
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Percentage format
pd.options.display.float_format = '${:,.2f}'.format

#Purhcasing analysis calculations for unique items, average purchase price, total number of purhcases, and total revenue
avg_purchase = purchase_data['Price'].mean()
numberOf_purchases = purchase_data['Item ID'].count()
uniques = len(purchase_data['Item ID'].unique())
tot_revenue = purchase_data['Price'].sum(axis=0)
#Print(avg_purchase,numberOf_purchases,uniques,tot_revenue)

#Create new Data Frame organized with the variables above 
purchasing_analysis = pd.DataFrame({'Total Unique Items':[uniques],
                                      'Average Purchase Price':[avg_purchase],
                                      'Total Number of Purchases':[numberOf_purchases],
                                      'Total Revenue':[tot_revenue]})
purchasing_analysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Percentage format
pd.options.display.float_format = '{:,.2f}%'.format

#Group items in the database by 'Gender' & 'Series Name' and count each time a 'Series Name' appears by 'Gender'
gender_demo_df = purchase_data.groupby('Gender')['SN'].nunique()
gender_demo_df

#Assign values to variables using slicing
female,male,other = gender_demo_df[0],gender_demo_df[1],gender_demo_df[2]

#Perform percentage calculations of each gender variable respectively
per_male,per_female,per_other = (male/players_count)*100,(female/players_count)*100,(other/players_count)*100

#Create new Data Frame organized with the variables above
demo_df = pd.DataFrame({'Total Players by Gender':[male,female,other],
                       'Percentage':[per_male,per_female,per_other]},
                      index=['Male','Female','Other'])
demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Accounting format
pd.options.display.float_format = '${:,.2f}'.format

#Create new Data Frame grouping original data by 'Gender' & 'Price'
purchasing_df = purchase_data.groupby('Gender')['Price']

#Perform counts on the number of times a purchase was made by each in 'Gender', get the average price of each purchase, sum all purchases, and get the average total purchase by each in 'Gender'
count = purchasing_df.count()
average = purchasing_df.mean()
tot_purchase = purchasing_df.sum()
avg_tot_purchase = tot_purchase/demo_df['Total Players by Gender']

#Create new Data Frame organized with the variables above and assign desired indices
new_purchasing_df = pd.DataFrame({'Purchase Count':[count[1],count[0],count[2]],
                                 'Average Purchase Price':[average[1],average[0],average[2]],
                                 'Total Purchase':[tot_purchase[1],tot_purchase[0],tot_purchase[2]],
                                 'Average Total Purchase':[avg_tot_purchase[1],avg_tot_purchase[0],avg_tot_purchase[2]]},
                                index=['Male','Female','Other'])
new_purchasing_df
```

```python
#Make a copy of the original data to avoid drawbacks
age_df = purchase_data.copy()

#Create a new Data Frame and assign the newly created 'age_df' to it. Group 'Series Names' with 'Age', count values and reset indices
grouped_df = age_df.groupby(['SN','Age']).count().reset_index()

#Create numerical age bins and a list containing series indices as strings
bins = [0,9.90,14,19.90,24.90,29.90,34.90,39.90,120]
series_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

#Cut data based on 'Age' and segment it into the age bins. Assing series_names as new indices on the Data Frame
grouped_df['Age ranges'] = pd.cut(x=grouped_df['Age'],bins=bins,labels=series_names)
grouped_df
```

```python
#Percentage Format
pd.options.display.float_format = '{:,.2f}%'.format

#Create a new Data Frame based on the cut DF and count values amongst it
age_demo = pd.DataFrame(grouped_df['Age ranges'].value_counts())

#Rename columns in the newly created Data Frame
age_demo.rename(columns={'Age ranges':'Total Count'},inplace=True)

#Create new Series in the Data Frame and assign percentage values based on calculations
age_demo['Percentage of Players'] = age_demo['Total Count']/players_count*100

#Reset your indices so that they are in descending order
age_demo = age_demo.reindex(index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Create a copy of the original data to avoid drawbacks and assign it into a new Data Frame
purchasing_analysis_df = purchase_data.copy()

#Accounting format
pd.options.display.float_format = '${:,.2f}'.format

#Create new serie: 'Age ranges' in the newly created Data Frame. Cut the data by 'Age' and segment it into age bins. Assign 'series_names' to labels
purchasing_analysis_df['Age ranges'] = pd.cut(purchasing_analysis_df['Age'],bins=bins,labels=series_names)

#Perform calculations to obtain purchase count, avg. purchase price, avg. purchase total per person and rename each series names respectively
avg_purchase_price = purchasing_analysis_df.groupby(['Age ranges']).mean()['Price'].rename('Average Purchase Price')
totalPurchase = purchasing_analysis_df.groupby(['Age ranges']).sum()['Price'].rename('Total Purchase Value')
purchaseCount = purchasing_analysis_df.groupby(['Age ranges']).count()['Price'].rename('Purchase Count')
totalAverage = totalPurchase/age_demo['Total Count']

#Create a new Data Frame organized with the variables above
purchasing_age_df = pd.DataFrame({'Purchase Count':purchaseCount,
                                 'Average Purchase Price':avg_purchase_price,
                                 'Total Purchase Value':totalPurchase,
                                 'Average Purchase per Person':totalAverage,})
purchasing_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Accounting Format
pd.options.display.float_format = '${:,.2f}'.format

#Group original data by 'Series Name'
summary_df = purchase_data.groupby('SN')

#Calculate number of purchases, avg. purchase price and total purchase value
numberOfPurchases = summary_df['Purchase ID'].count()
purchaseValue = summary_df['Price'].sum()
avgPurchaseValue = summary_df['Price'].mean()

#Create new Data Frame organized with the variables above
top_spenders_df = pd.DataFrame({'Purchase Count':numberOfPurchases,
                               'Average Purchase Price':avgPurchaseValue,
                               'Total Purchase Value':purchaseValue})

#Sort the 'Total Purchase Value' series in descending order
formatted_top_spenders_df = top_spenders_df.sort_values(['Total Purchase Value'],ascending=False)
formatted_top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Accounting format
pd.options.display.float_format = '${:,.2f}'.format

#Create a copy of the original data to avoid drawbacks
popular_df = purchase_data.copy()

#Slice desired columns
popularItems = popular_df[['Item ID','Item Name','Price']]

#Group desired columns and perform calculations to obtain purchase count, item price, and total purchase value
grouped_items = popularItems.groupby(['Item ID','Item Name'])
priceSum = grouped_items['Price'].sum()
newPurchaseCount = grouped_items['Price'].count()
pricePerItem = priceSum/newPurchaseCount

#Create a new Data Frame using variables created above
popular_items_df = pd.DataFrame({'Purchase Count':newPurchaseCount,
                                'Item Price':pricePerItem,
                                'Total Purchase Value':priceSum},)

#Sort values in the 'Purchase Count' series in descending order
most_popular_df = popular_items_df.sort_values(['Purchase Count'],ascending=False)
most_popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort values in the 'Total Purchase Value' series in descending order
most_profitable_df = popular_items_df.sort_values(['Total Purchase Value'],ascending=False)
most_profitable_df.head()
```
